## K-Fold

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.datasets import make_regression
```

```python
# Convert the continuous target into bins
n_bins = 6  # you can adjust this number
y_binned = np.floor(
    np.interp(y, (y.min(), y.max()), (0, n_bins))
    ).astype(int)
```

```python
# Create StratifiedKFold with shuffle and seed
skf = StratifiedKFold(n_splits=n_bins, shuffle=True, random_state=1360)
```

```python
# Use the binned labels for stratification
for fold, (train_idx, test_idx) in enumerate(skf.split(x, y_binned)):
    x_train, x_test = x[train_idx], x[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    print(f"Fold {fold+1}:")
    print(f"  Train size: {len(train_idx)}, Val size: {len(test_idx)}")
```

```python
plot_xy([x_train, y_train])
```

```python
# Crear una figura con dos subplots en horizontal (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # figsize ajusta el tamaño

bin_edges = np.linspace(y.min(), y.max(), n_bins + 1)

# Graficar la primera curva en el primer subplot
ax1.hist(y_train, color='blue', label='train', bins = bin_edges, edgecolor='black')
ax1.set_title('Train Density')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Graficar la segunda curva en el segundo subplot
ax2.hist(y_test, color='red', label='test', bins = bin_edges, edgecolor='black')
ax2.set_title('Test Density')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
# Get bin edges using linspace
bin_edges = np.linspace(y.min(), y.max(), n_bins + 1)

# Plot histogram using the same bin edges
plt.hist(y_train, bins=bin_edges, edgecolor='black')
plt.hist(y_test, bins=bin_edges, edgecolor='black')
plt.xlabel("Target value")
plt.ylabel("Count")
plt.title("Histogram of Target Values (Binned for Stratified K-Fold)")
plt.grid(True)
plt.show()
```

## Training Session

```python
from sklearn.linear_model import LinearRegression
```

```python
mape = lambda y_true, y_pred:  100 * np.mean(np.abs( (y_true - y_pred)/(y_true)  ))
```

```python
import warnings

warnings.filterwarnings("ignore", message=".*least populated class in y has only.*")
```

```python
_results = []
# Use the binned labels for stratification
for fold, (train_idx, test_idx) in enumerate(skf.split(x, y_binned)):
    x_train, x_test = x[train_idx], x[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    print(f"Fold {fold+1}:")
    print(f"  Train size: {len(train_idx)}, Val size: {len(test_idx)}")

    model = LinearRegression()
    model.fit(x_train, y_train)

    # Get coefficients and bias
    coefficients = model.coef_  # Array of 3 coefficients (one for each x variable)
    bias = model.intercept_    # The bias term

    print("R2: ", model.score(x_test, y_test))
    print("MAPE", mape(y_test, model.predict(x_test)) )
    values = [ float(j) for j in model.coef_ ] + [float(bias)] + [model.score(x_test, y_test)] + [float(mape(y_test, model.predict(x_test)))]

    _results.append(values)

```

```python
_results = np.array(_results)
_results.shape
```

```python
print('Avg MAPE: {:.2f} - STD {:.2f}'.format(np.mean(_results[:, -1]), np.std(_results[:, -1]) ))
print('Avg R2: {:.2f} - STD {:.2f}'.format(np.mean(_results[:, -2]), np.std(_results[:, -2]) ))
```

```python
# Crear una figura con dos subplots en horizontal (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # figsize ajusta el tamaño


# Graficar la primera curva en el primer subplot
ax1.plot(_results[:, -1], color='blue', label='MAE')
ax1.set_title('Train')
ax1.set_xlabel('Fold')
ax1.set_ylabel('MAE')
ax1.grid()
ax1.legend()

# Graficar la segunda curva en el segundo subplot
ax2.plot(_results[:, -2], color='red', label=r'$R^{2}$')
ax2.set_title('Test')
ax2.set_xlabel('Fold')
ax2.set_ylabel(r'$R^{2}$')
ax2.grid()
ax2.legend()

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```

## K-Fold Results

## Used libraries

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

## Load the KFold Results

```python
data = []
for i in range(6): 
    data.append( np.load('./kfolds_reg/{}.npy'.format(i))[0] )
```

```python
data = np.array(data)
data
```

```python
data.shape
```

## Plots

```python
# Create figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns
x = [x for x in range(1, 7)]

# First plot
ax1.plot(x, data[:, -2], color='blue')
ax1.set_title('Cost Function')
ax1.set_xlabel('K-Fold')
ax1.set_ylabel('MAE')

# Second plot
ax2.plot(x, data[:, -1], color='red')
ax2.set_title('Determination Coefficiente')
ax2.set_xlabel('K-Fold')
ax2.set_ylabel(r'$R^{2}$')

plt.tight_layout()
plt.show()
```

```python
print( 'MAE: {:.2f} STD: {:.2f} - HB: {:.2f}'.format(
    np.mean(data[:, -2]), 
    np.std(data[:, -2]), 
    (np.mean(data[:, -2]) + np.std(data[:, -2])) 
     ) )
```

```python
print( 'R2: {:.2f} STD: {:.2f} - LB: {:.2f}'.format(
    np.mean(data[:, -1]), 
    np.std(data[:, -1]), 
    (np.mean(data[:, -1]) - np.std(data[:, -1])) 
))
```
